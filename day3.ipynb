```python
with open('3a2.txt') as f:
    lines = f.readlines()
```

### Part 1

```python
gamma = []
epsilon = []

for i in range(len(lines[0])-1):
    zeroCount = 0
    oneCount = 0
    for line in lines:
        if line[i] == '0': zeroCount += 1
        elif line[i] == '1': oneCount += 1
    if zeroCount < oneCount:
        gamma.append('0')
        epsilon.append('1')
    else:
        gamma.append('1')
        epsilon.append('0')

gammaStr = "".join(gamma) 
epsilonStr = "".join(epsilon)
# now we have to convert gamma & epsilon to decimals
gammaDec = int(gammaStr,2)
epsilonDec = int(epsilonStr,2)

print(epsilonStr)
print(gammaStr)
print(gammaDec * epsilonDec)

# print(len(lines))

part1(lines)
```
