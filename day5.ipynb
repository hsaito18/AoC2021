```python
with open('5a.txt') as f:
    lines = f.readlines()
```

```python
import re
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)
def parseLine(line):
    no_whitespace = ''.join(line.split())
    return re.split(',|->', no_whitespace)
```

### Part 1

```python
straightLines = []
diagonalLines = []
for line in lines:
    coords = parseLine(line)
    coords = [int(item) for item in coords]
    if coords[0] == coords[2] and coords[1] != coords[3] or coords[1] == coords[3] and coords[0] != coords[2]: straightLines.append(coords)
    else: diagonalLines.append(coords)


grid = np.zeros((1000,1000))

for line in straightLines:
    if line[0] == line[2]:
        startCol = min(line[1], line[3])
        stopCol = max(line[1], line[3])
        for col in range(startCol, stopCol + 1):
            grid[line[0]][col] += 1
    elif line[1] == line[3]:
        startRow = min(line[0], line[2])
        stopRow = max(line[0], line[2])
        for row in range(startRow, stopRow + 1):
            grid[row][line[1]] += 1

sum = 0
for r in range(1000):
    for c in range(1000):
       if grid[r][c] > 1:
           sum += 1

print(sum)
# 7638 too low.
 
```

### Part 2

```python
grid = np.zeros((1000,1000))

for line in straightLines:
    if line[0] == line[2]:
        startCol = min(line[1], line[3])
        stopCol = max(line[1], line[3])
        for col in range(startCol, stopCol + 1):
            grid[line[0]][col] += 1
    elif line[1] == line[3]:
        startRow = min(line[0], line[2])
        stopRow = max(line[0], line[2])
        for row in range(startRow, stopRow + 1):
            grid[row][line[1]] += 1

for line in diagonalLines:
    row1 = line[0]
    row2 = line[2]
    col1 = line[1]
    col2 = line[3]
    diff = np.abs(row1-row2)
    row_dir = np.sign(row2-row1)
    col_dir = np.sign(col2-col1)
    for i in range(0, diff+1):
        grid[row1 + i * row_dir][col1 + i * col_dir] += 1
    
sum = 0
for r in range(1000):
    for c in range(1000):
       if grid[r][c] > 1:
           sum += 1

print(sum)
print(np.transpose(grid))
```
